## Target: Predict the number of bicycle rentals for a given day. - Casual + Registered
- Overview: https://byui-cse.github.io/cse450-course/module-04/
- Data Dictionary: https://byui-cse.github.io/cse450-course/module-04/bikes-dictionary.txt
- Data: https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bikes.csv

```python
import pandas as pd
bike_rental_data = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bikes.csv')
bike_rental_data.describe()

#hour holiday, working a

```

```python
# Adds the registed and casual together into a new column called total rentals:
bike_rental_data['total_rentals'] = bike_rental_data['casual'] + bike_rental_data['registered']

# Drops the registered and casual columns
bike_rental_data = bike_rental_data.drop(['registered', 'casual'], axis=1)
```

## Holiday: 

```python
import altair as alt
sampled_data = bike_rental_data.sample(n=1000, random_state=42)  # Sample 1000 random rows from the dataset

aggregated_data = bike_rental_data.groupby('holiday').agg({'total_rentals': 'sum'}).reset_index()

bar_chart = alt.Chart(aggregated_data).mark_bar().encode(
    x='holiday:N',
    y='total_rentals:Q',
    color='holiday:N'
).properties(
    width=500,
    height=300,
    title='Bar Chart of Holiday vs. Total Rentals (Aggregated Data)'
)

bar_chart



```

## Hour: 17, 18, 8, 19, 16 - Keep

```python
import altair as alt
import pandas as pd

# Aggregate data by hour and calculate the total rentals
hourly_data = bike_rental_data.groupby('hr').agg({'total_rentals': 'sum'}).reset_index()

# Sort the data by the 'total_rentals' column in ascending order
hourly_data = hourly_data.sort_values('total_rentals')

# Create the bar chart
bar_chart = alt.Chart(hourly_data).mark_bar().encode(
    x=alt.X('hr:O', title='Hour'),
    y=alt.Y('total_rentals:Q', title='Total Rentals')
).properties(
    width=500,
    height=300,
    title='Bar Chart of Hour vs. Total Rentals (Sorted)'
)

bar_chart
```

## Working Days - Drop

```python
import pandas as pd
import altair as alt

aggregate_data = bike_rental_data.groupby('workingday')['total_rentals'].mean().reset_index()

chart = alt.Chart(aggregate_data).mark_bar().encode(
    x='workingday',
    y='total_rentals',
    color=alt.Color('workingday', legend=None),
    tooltip=['workingday', 'total_rentals']
).properties(
    title='Average Rentals per Working Day'
)

chart

# bike_rental_data['dteday'].value_counts()
bike_rental_data.drop('workingday', axis=1)
```
